## Importing libraries and data

```python
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
# loading dataset 
data = sns.load_dataset("iris") 
```

```python
data.head()
```

```python
data.shape
```

## Simple plot

```python
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data);
```

## Using Seaborn with Matplotlib

```python
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) ;
  
# setting the title using Matplotlib
plt.title('Title using Matplotlib Function')
  
plt.show()
```

### Setting the xlim and ylim

```python
# loading dataset 
data = sns.load_dataset("iris") 
  
# draw lineplot 
sns.lineplot(x="sepal_length", y="sepal_width", data=data) ;
  
# setting the x limit of the plot
plt.xlim(4, 7)
plt.ylim(2.5, 3.8)

plt.show()
```

## Relational Plots

### scatter plot

```python
sns.scatterplot(x='species', y='sepal_width', data=data)
  
plt.show()
```

```python
sns.scatterplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

### Line Plot

```python
sns.lineplot(x='sepal_length', y='species', data=data)
plt.show()
```

## Categorical Plots

### Bar Plot

```python
sns.barplot(x='species', y='sepal_length', data=data)
plt.show()
```

### Count Plot

```python
sns.countplot(x='species', data=data)
plt.show()
```

### Box Plot

```python
sns.boxplot(x='species', y='sepal_width', data=data)
plt.show()
```

### Violinplot

```python
sns.violinplot(x='species', y='sepal_width', data=data)
plt.show()
```

### Stripplot

```python
sns.stripplot(x='species', y='sepal_width', data=data)
plt.show()
```

## Distribution Plots

### Histogram

```python
sns.histplot(x='species', y='sepal_width', data=data)
plt.show()
```

### Jointplot

```python
sns.jointplot(x='sepal_length', y='sepal_width', data=data)
plt.show()
```

### Pairplot

```python
sns.pairplot(data=data, hue='species')
plt.show()
```

## Matrix Plots

```python
# correlation between the different parameters 
c = data.corr()
  
sns.heatmap(c)
plt.show()
```
