## Format Strings fmt

`marker|line|color`

- `o:r`
    - `o`  -> circle marker
    - `:`  -> dotted line
    - `r`  -> red color
    
- `X--g`
    - `X`  -> filled x marker
    - `--` -> dashed line
    - `g`  -> green color
    
- `+-.y`
    - `+`  -> plus marker
    - `-.` -> dashed dotted line
    - `y`  -> yellow color    

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o:r')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'X--g')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, '+-.y')
plt.show()
```

## Marker Size
- `markersize` or shorter `ms`

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, 'o-.y', markersize = 20)
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20)
plt.show()
```

## Marker Color

### marker edge color
- `markeredgecolor` or shorter `mec`

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, markeredgecolor = 'r')
plt.show()
```

### marker face color
- `markerfacecolor` or shorter `mfc`

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'X', ms = 20, markerfacecolor = 'r')
plt.show()
```

### marker edge color and marker face color

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc = 'r')
plt.show()
```

## Linestyle
- `linestyle` or the shorter `ls`

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dotted')
plt.show()
```

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linestyle = 'dashed')
plt.show()
```

### Shorter Syntax

- The line style can be written in a shorter syntax:

    - `linestyle` can be written as `ls`.

    - `dotted` can be written as `:`

    - `dashed` can be written as `--`.

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, ls = ':')
plt.show()
```

### Line Color
- `color` or the shorter `c`

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, color = 'r')
plt.show()
```

### Line Width
- `linewidth` or shorter `lw`

```python
ypoints = np.array([3, 8, 1, 10])

plt.plot(ypoints, linewidth = '20.5')
plt.show()
```

## Labels and Title

### Create Labels for a Plot

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### Create a Title for a Plot

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

### Set Font Properties for Title and Labels
- we can use the `fontdict` parameter in `xlabel()`, `ylabel()`, and `title()` to set font properties for the title and labels.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

### Position the Title

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data", loc = 'right')
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

## Display Multiple Plots


- The subplot() function takes three arguments that describes the layout of the figure.

- The layout is organized in rows and columns, which are represented by the first and second argument.

- The third argument represents the index of the current plot.
</br>


`plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
`
</br>
</br>

`
plt.subplot(1, 2, 2)
#the figure has 1 row, 2 columns, and this plot is the second plot.
`


```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 2)
plt.plot(x,y)

#plot 3:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 2, 3)
plt.plot(x,y)

#plot 4:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 2, 4)
plt.plot(x,y)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```
