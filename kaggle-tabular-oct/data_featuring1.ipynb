## Eliminando columnas no útiles

```python
import gc
gc.collect()
from lightgbm import LGBMClassifier
#df.dropna(inplace = True)
y_b = df['team_B_scoring_within_10sec']
y_a = df['team_A_scoring_within_10sec']
to_drop = ['game_num','event_id', 'event_time', 'player_scoring_next',
       'team_scoring_next',
       'team_B_scoring_within_10sec','team_A_scoring_within_10sec']
df.drop(to_drop, inplace = True, axis = 1)
```

```python
df.keys()
```

```python
df['p0_norm']=(df['p0_pos_x']**2+df['p0_pos_y']**2+df['p0_pos_z']**2 )**.5
df['p1_norm']=(df['p1_pos_x']**2+df['p1_pos_y']**2+df['p1_pos_z']**2 )**.5
df['p2_norm']=(df['p2_pos_x']**2+df['p2_pos_y']**2+df['p2_pos_z']**2 )**.5
df['p3_norm']=(df['p3_pos_x']**2+df['p3_pos_y']**2+df['p3_pos_z']**2 )**.5
df['p4_norm']=(df['p4_pos_x']**2+df['p4_pos_y']**2+df['p4_pos_z']**2 )**.5
df['p5_norm']=(df['p5_pos_x']**2+df['p5_pos_y']**2+df['p5_pos_z']**2 )**.5
```

```python
df['p0_norm'].isna().sum()
```

```python
df['p0_off'] = 1-df.loc[:,['p0_norm']].count(axis=1)
df['p1_off'] = 1-df.loc[:,['p1_norm']].count(axis=1)
df['p2_off'] = 1-df.loc[:,['p2_norm']].count(axis=1)
df['p3_off'] = 1-df.loc[:,['p3_norm']].count(axis=1)
df['p4_off'] = 1-df.loc[:,['p4_norm']].count(axis=1)
df['p5_off'] = 1-df.loc[:,['p5_norm']].count(axis=1)
```

```python
df['teamA_off']=df['p0_off']+df['p1_off']+df['p2_off'] 
df['teamB_off']=df['p3_off']+df['p4_off']+df['p5_off']
df 
```

```python
to_drop = ['p0_off','p1_off', 'p2_off', 'p3_off',
       'p4_off', 'p5_off']
df.drop(to_drop, inplace = True, axis = 1)
df
```

```python
#Train
model = LGBMClassifier()
model.fit(df, y_a)
print('Model A trained')
```
