```python
df['normA'] = (df['A1']**2+df['A2']**2)**.5
df['normB'] = (df['B1']**2+df['B2']**2)**.5
df
```

```python
df['A_off'] = 1-df.loc[:,['normA']].count(axis=1)
df['B_off'] = 1-df.loc[:,['normB']].count(axis=1)
df
```

```python
df['team1_off']=df['A_off']+df['B_off']
df
```

```python
df['A1'] = df['A1'].fillna(-100)
df['A2'] = df['A2'].fillna(-100)
df['B1'] = df['B1'].fillna(-100)
df['B2'] = df['B2'].fillna(-100)
df
```
